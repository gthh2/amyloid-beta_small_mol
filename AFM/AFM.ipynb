This notebook describes the analysis of Atomic Force Microscopy (AFM) experiments of amyloid beta (AB) fibrils formed in the absence and presence the small molecule, 10074-G5.
Fibrils were formed using 1 uM amyloid beta in the absence and presence of 6 uM 10074-G5 at 37 degrees C.

```python
#import useful things
import numpy as np
import pandas as pd
import math
import seaborn as sns
import matplotlib.pyplot as plt; plt.rcdefaults()
%matplotlib inline
```

```python
#load in data
AB_2_5hours = pd.read_csv('amyloid_beta_2_5hours.csv',  header=None).values.flatten()
G5_2_5hours = pd.read_csv('10074G5_2_5hours.csv',  header=None).values.flatten()

AB_7_5hours = pd.read_csv('amyloid_beta_7_5hours.csv',  header=None).values.flatten()
G5_7_5hours = pd.read_csv('10074G5_7_5hours.csv',  header=None).values.flatten()
```

```python
#make pandas data frame for easy plotting
data = {'AB_2_5hours': AB_2_5hours, 'G5_2_5hours': G5_2_5hours, 'AB_7_5hours': AB_7_5hours, 'G5_7_5hours': G5_7_5hours}
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in data.items() ]), columns=['AB_2_5hours','G5_2_5hours','AB_7_5hours','G5_7_5hours'])
df.columns = [list(['AB 2.5 hours', 'G5 2.5 hours', 'AB 7.5 hours', 'G5 7.5 hours'])]
df.columns.names = ['Sample']
df_melt = pd.melt(df) 
```

```python
#make strip plot 
ax = sns.stripplot(data=df_melt, x='Sample', y='value', jitter=True)
plt.ylabel("Fibril height (nm)")
plt.show()
```
