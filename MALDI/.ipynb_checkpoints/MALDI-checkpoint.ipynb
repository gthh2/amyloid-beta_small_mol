This notebook describes the matrix-assisted laser desorption/ionization (MALDI) mass spectrometry data of of monomeric amyloid beta 42 (AB) in the presence and absence of the small molecule, 10074-G5.

```python
#import useful things
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt; plt.rcdefaults()
%matplotlib inline
```

```python
#load in data
Mass                    = pd.read_csv('Mass_m_over_z.csv',                 header=None).values.flatten()
AB_alone_sup            = pd.read_csv('AB42_alone_supernatant.csv',        header=None).values.flatten()
AB_10074G5_sup          = pd.read_csv('AB42_with_10074G5_supernatant.csv', header=None).values.flatten()
AB_alone_pel            = pd.read_csv('AB42_alone_pellet.csv',             header=None).values.flatten()
AB_10074G5_pel          = pd.read_csv('AB42_with_10074G5_pellet.csv',      header=None).values.flatten()
```

```python
#Plot AB alone, supernatant
plt.scatter(Mass, AB_alone_sup, s=0.05)
plt.title("AB alone, supernatant")
plt.xlabel("Mass (m/z)")
plt.ylabel("Intensity")
plt.show()
```

```python
#Plot AB with 10074-G5, supernatant
plt.scatter(Mass, AB_10074G5_sup, s=0.05)
plt.title("AB with 10074-G5, supernatant")
plt.xlabel("Mass (m/z)")
plt.ylabel("Intensity")
plt.show()
```

```python
#Plot AB alone, pellet
plt.scatter(Mass, AB_alone_pel, s=0.05)
plt.title("AB alone, pellet")
plt.xlabel("Mass (m/z)")
plt.ylabel("Intensity")
plt.show()
```

```python
#Plot AB with 10074-G5, pellet
plt.scatter(Mass, AB_10074G5_pel, s=0.05)
plt.title("AB with 10074-G5, pellet")
plt.xlabel("Mass (m/z)")
plt.ylabel("Intensity")
plt.show()
```
