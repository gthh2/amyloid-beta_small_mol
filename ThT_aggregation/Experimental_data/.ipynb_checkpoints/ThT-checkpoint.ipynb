This notebook shows aggregation data collected using the fluorescent dye, Thioflavin T (ThT). See 'Kinetic_modelling' directory for details on the mathematical fitting and Mathematica notebooks.

```python
#import useful things
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt; plt.rcdefaults()
%matplotlib inline
```

```python
#load in data from '1uM_AB42_varying_10074_G5' 
Time_1 = pd.read_csv('1uM_AB42_varying_10074_G5/Time_mins.csv',  header=None).values.flatten()

AB42_1uM_10074G5_0uM = pd.read_csv('1uM_AB42_varying_10074_G5/0uM_10074_G5.csv',  header=None).values
AB42_1uM_10074G5_1uM = pd.read_csv('1uM_AB42_varying_10074_G5/1uM_10074_G5.csv',  header=None).values
AB42_1uM_10074G5_2uM = pd.read_csv('1uM_AB42_varying_10074_G5/2uM_10074_G5.csv',  header=None).values
AB42_1uM_10074G5_6uM = pd.read_csv('1uM_AB42_varying_10074_G5/6uM_10074_G5.csv',  header=None).values
AB42_1uM_10074G5_10uM = pd.read_csv('1uM_AB42_varying_10074_G5/10uM_10074_G5.csv',  header=None).values
AB42_1uM_10074G5_15uM = pd.read_csv('1uM_AB42_varying_10074_G5/15uM_10074_G5.csv',  header=None).values
AB42_1uM_10074G5_20uM = pd.read_csv('1uM_AB42_varying_10074_G5/20uM_10074_G5.csv',  header=None).values
```

```python
#Plot of the ThT data with 1uM AB42 and varying 10074-G5 concentrations
plt.errorbar(Time_1, AB42_1uM_10074G5_0uM.mean(axis=1),  yerr=AB42_1uM_10074G5_0uM.std(axis=1),  ms=4, color='grey',   ecolor='lightgrey',   marker=".", label='0 uM')
plt.errorbar(Time_1, AB42_1uM_10074G5_1uM.mean(axis=1),  yerr=AB42_1uM_10074G5_1uM.std(axis=1),  ms=4, color='red',    ecolor='lightcoral',    marker=".", label='1 uM')
plt.errorbar(Time_1, AB42_1uM_10074G5_2uM.mean(axis=1),  yerr=AB42_1uM_10074G5_2uM.std(axis=1),  ms=4, color='orange', ecolor='bisque',  marker=".", label='2 uM')
plt.errorbar(Time_1, AB42_1uM_10074G5_6uM.mean(axis=1),  yerr=AB42_1uM_10074G5_6uM.std(axis=1),  ms=4, color='yellow', ecolor='lightyellow', marker=".", label='6 uM')
plt.errorbar(Time_1, AB42_1uM_10074G5_10uM.mean(axis=1), yerr=AB42_1uM_10074G5_10uM.std(axis=1), ms=4, color='green',  ecolor='lightgreen',  marker=".", label='10 uM')
plt.errorbar(Time_1, AB42_1uM_10074G5_15uM.mean(axis=1), yerr=AB42_1uM_10074G5_15uM.std(axis=1), ms=4, color='blue',   ecolor='lightblue',   marker=".", label='15 uM')
plt.errorbar(Time_1, AB42_1uM_10074G5_20uM.mean(axis=1), yerr=AB42_1uM_10074G5_20uM.std(axis=1), ms=4, color='purple', ecolor='thistle', marker=".", label='20 uM')

plt.title("1 uM AB42 and varying 10074-G5 concentrations")
plt.xlabel("Time (mins)")
plt.ylabel("Fluorescence intensity")
plt.legend(loc="upper left")
plt.show()
```

```python
#load in data from 'varying_AB42_0uM_10074_G5'
Time_2 = pd.read_csv('varying_AB42_0uM_10074_G5/Time_mins.csv',  header=None).values.flatten()

cont_AB42_1uM_10074G5_0uM = pd.read_csv('varying_AB42_0uM_10074_G5/1uM_AB.csv',  header=None).values
cont_AB42_1_5uM_10074G5_0uM = pd.read_csv('varying_AB42_0uM_10074_G5/1_5uM_AB.csv',  header=None).values
cont_AB42_2uM_10074G5_0uM = pd.read_csv('varying_AB42_0uM_10074_G5/2uM_AB.csv',  header=None).values
```

```python
#Plot of the ThT data with 0uM 10074-G5 and varying AB42 concentrations 
plt.errorbar(Time_2, cont_AB42_1uM_10074G5_0uM.mean(axis=1),  yerr=cont_AB42_1uM_10074G5_0uM.std(axis=1), ms=4, color='lightgrey',   marker=".", label='1 uM')
plt.errorbar(Time_2, cont_AB42_1_5uM_10074G5_0uM.mean(axis=1),  yerr=cont_AB42_1_5uM_10074G5_0uM.std(axis=1), ms=4, color='grey',   marker=".", label='1.5 uM')
plt.errorbar(Time_2, cont_AB42_2uM_10074G5_0uM.mean(axis=1),    yerr=cont_AB42_2uM_10074G5_0uM.std(axis=1),   ms=4, color='black', marker=".", label='2 uM')

plt.title("0uM 10074-G5 and varying AB42 concentrations")
plt.xlabel("Time (mins)")
plt.ylabel("Fluorescence intensity")
plt.legend(loc="upper left")
plt.show()
```

```python
#Plot of the ThT elongation data with 0.85 uM monomeric AB42 and 0.15 uM preformed fibril seeds and varying 10074-G5 concentrations
plt.errorbar(Time_4, elong_AB42_1uM_10074G5_0uM.mean(axis=1),  yerr=elong_AB42_1uM_10074G5_0uM.std(axis=1),  ms=4, color='grey',   ecolor='lightgrey',   marker=".", label='0 uM')
plt.errorbar(Time_4, elong_AB42_1uM_10074G5_1uM.mean(axis=1),  yerr=elong_AB42_1uM_10074G5_1uM.std(axis=1),  ms=4, color='red',    ecolor='lightcoral',    marker=".", label='1 uM')
plt.errorbar(Time_4, elong_AB42_1uM_10074G5_2uM.mean(axis=1),  yerr=elong_AB42_1uM_10074G5_2uM.std(axis=1),  ms=4, color='orange', ecolor='bisque',  marker=".", label='2 uM')
plt.errorbar(Time_4, elong_AB42_1uM_10074G5_6uM.mean(axis=1),  yerr=elong_AB42_1uM_10074G5_6uM.std(axis=1),  ms=4, color='yellow', ecolor='lightyellow', marker=".", label='6 uM')
plt.errorbar(Time_4, elong_AB42_1uM_10074G5_10uM.mean(axis=1), yerr=elong_AB42_1uM_10074G5_10uM.std(axis=1), ms=4, color='green',  ecolor='lightgreen',  marker=".", label='10 uM')
plt.errorbar(Time_4, elong_AB42_1uM_10074G5_20uM.mean(axis=1), yerr=elong_AB42_1uM_10074G5_20uM.std(axis=1), ms=4, color='purple', ecolor='thistle', marker=".", label='20 uM')

plt.title("1 uM AB42 (15% seeds) and varying 10074-G5 concentrations")
plt.xlabel("Time (mins)")
plt.ylabel("Fluorescence intensity")
plt.legend(loc="upper left")
plt.show()
```
