This notebook shows data collected from Bio-Layer Interferometry (BLI) experiments of monomeric amyloid beta binding to various concetrations of the small molecule, 10074-G5.

Monomeric amyloid beta (AB) was functionalised on Super Streptavidin biosensors and repeatedly exposed to various solutions of 10074-G5 and buffer (repeated six times). Control biosensors were functionalied with biocytin to test for non-specific binding.

```python
#import useful things
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt; plt.rcdefaults()
%matplotlib inline
```

```python
#load in data
Time = pd.read_csv('amyloid_beta_tip/Time.csv',  header=None).values.flatten()

AB_1uM = pd.read_csv('amyloid_beta_tip/1_uM.csv',  header=None).values.flatten()
biocytin_1uM = pd.read_csv('biocytin_control_tip/1uM.csv',  header=None).values.flatten()

AB_2uM = pd.read_csv('amyloid_beta_tip/2_uM.csv',  header=None).values.flatten()
biocytin_2uM = pd.read_csv('biocytin_control_tip/2uM.csv',  header=None).values.flatten()

AB_3uM = pd.read_csv('amyloid_beta_tip/3_uM.csv',  header=None).values.flatten()
biocytin_3uM = pd.read_csv('biocytin_control_tip/3uM.csv',  header=None).values.flatten()

AB_4uM = pd.read_csv('amyloid_beta_tip/4_uM.csv',  header=None).values.flatten()
biocytin_4uM = pd.read_csv('biocytin_control_tip/4uM.csv',  header=None).values.flatten()

AB_5uM = pd.read_csv('amyloid_beta_tip/5_uM.csv',  header=None).values.flatten()
biocytin_5uM = pd.read_csv('biocytin_control_tip/5uM.csv',  header=None).values.flatten()
```

```python
#Plot of the real time data using amyloid beta functionalised biosensors
plt.scatter(Time, AB_1uM, s=0.5, color='red',    label='1 uM')
plt.scatter(Time, AB_2uM, s=0.5, color='orange', label='2 uM')
plt.scatter(Time, AB_3uM, s=0.5, color='yellow', label='3 uM')
plt.scatter(Time, AB_4uM, s=0.5, color='green',  label='4 uM')
plt.scatter(Time, AB_5uM, s=0.5, color='blue',   label='5 uM')
plt.title("Amyloid beta tip")
plt.xlabel("Time (mins)")
plt.ylabel("BLI signal")
plt.legend(loc="upper right")
plt.ylim(-.04, 0.12)
plt.show()
```

```python
#Plot of the real time data using biocytin control functionalised biosensors
plt.scatter(Time, biocytin_1uM, s=0.5, color='red',    label='1 uM')
plt.scatter(Time, biocytin_2uM, s=0.5, color='orange', label='2 uM')
plt.scatter(Time, biocytin_3uM, s=0.5, color='yellow', label='3 uM')
plt.scatter(Time, biocytin_4uM, s=0.5, color='green',  label='4 uM')
plt.scatter(Time, biocytin_5uM, s=0.5, color='blue',   label='5 uM')
plt.title("Biocytin control tip")
plt.xlabel("Time (mins)")
plt.ylabel("BLI signal")
plt.legend(loc="upper right")
plt.ylim(-.04, 0.12)
plt.show()
```

```python
#Split curves into individual replicates
Time_reps = np.split(Time, 6)

AB_1uM_reps = np.split(AB_1uM, 6)
AB_2uM_reps = np.split(AB_2uM, 6)
AB_3uM_reps = np.split(AB_3uM, 6)
AB_4uM_reps = np.split(AB_4uM, 6)
AB_5uM_reps = np.split(AB_5uM, 6)

biocytin_1uM_reps = np.split(biocytin_1uM, 6)
biocytin_2uM_reps = np.split(biocytin_2uM, 6)
biocytin_3uM_reps = np.split(biocytin_3uM, 6)
biocytin_4uM_reps = np.split(biocytin_4uM, 6)
biocytin_5uM_reps = np.split(biocytin_5uM, 6)
```

```python
#Plot of averaged data using amyloid beta functionalised biosensors
plt.errorbar(Time_reps[0], np.asarray(AB_1uM_reps).mean(axis=0), yerr=np.asarray(AB_1uM_reps).std(axis=0), capsize=None, color='red',    ecolor='mistyrose',   label='1 uM')
plt.errorbar(Time_reps[0], np.asarray(AB_2uM_reps).mean(axis=0), yerr=np.asarray(AB_2uM_reps).std(axis=0), capsize=None, color='orange', ecolor='bisque',      label='2 uM')
plt.errorbar(Time_reps[0], np.asarray(AB_3uM_reps).mean(axis=0), yerr=np.asarray(AB_3uM_reps).std(axis=0), capsize=None, color='yellow', ecolor='lightyellow', label='3 uM')
plt.errorbar(Time_reps[0], np.asarray(AB_4uM_reps).mean(axis=0), yerr=np.asarray(AB_4uM_reps).std(axis=0), capsize=None, color='green',  ecolor='lightgreen',  label='4 uM')
plt.errorbar(Time_reps[0], np.asarray(AB_5uM_reps).mean(axis=0), yerr=np.asarray(AB_5uM_reps).std(axis=0), capsize=None, color='blue',   ecolor='lightblue',   label='5 uM')
plt.title("Average response, amyloid beta tip")
plt.xlabel("Time (mins)")
plt.ylabel("BLI signal")
plt.legend(loc="upper right")
plt.ylim(-.04, 0.12)
plt.show()
```

```python
#Plot of averaged data using control biocytin functionalised biosensors
plt.errorbar(Time_reps[0], np.asarray(biocytin_1uM_reps).mean(axis=0), yerr=np.asarray(biocytin_1uM_reps).std(axis=0), capsize=None, color='red',    ecolor='mistyrose',   label='1 uM')
plt.errorbar(Time_reps[0], np.asarray(biocytin_2uM_reps).mean(axis=0), yerr=np.asarray(biocytin_2uM_reps).std(axis=0), capsize=None, color='orange', ecolor='bisque',      label='2 uM')
plt.errorbar(Time_reps[0], np.asarray(biocytin_3uM_reps).mean(axis=0), yerr=np.asarray(biocytin_3uM_reps).std(axis=0), capsize=None, color='yellow', ecolor='lightyellow', label='3 uM')
plt.errorbar(Time_reps[0], np.asarray(biocytin_4uM_reps).mean(axis=0), yerr=np.asarray(biocytin_4uM_reps).std(axis=0), capsize=None, color='green',  ecolor='lightgreen',  label='4 uM')
plt.errorbar(Time_reps[0], np.asarray(biocytin_5uM_reps).mean(axis=0), yerr=np.asarray(biocytin_5uM_reps).std(axis=0), capsize=None, color='blue',   ecolor='lightblue',   label='5 uM')
plt.title("Average response, biocytin control tip")
plt.xlabel("Time (mins)")
plt.ylabel("BLI signal")
plt.legend(loc="upper right")
plt.ylim(-.04, 0.12)
plt.show()
```
