This notebook describes the isothermal titration calorimetry (ITC) data of of monomeric amyloid beta 40 (AB40) titrated into a solution of the small molecule, 10074-G5.

```python
#import useful things
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt; plt.rcdefaults()
%matplotlib inline
```

```python
#load in data
Time              = pd.read_csv('Time.csv',               header=None).values.flatten()
AB40_to_10074G5   = pd.read_csv('AB40_to_10074G5.csv',    header=None).values.flatten()
AB40_to_buffer    = pd.read_csv('AB40_to_buffer.csv',     header=None).values.flatten()
Buffer_to_10074G5 = pd.read_csv('Buffer_to_10074G5.csv',  header=None).values.flatten()
Molar_ratio       = pd.read_csv('Molar_ratio.csv',        header=None).values.flatten()
dQ                = pd.read_csv('dQ.csv',                 header=None).values.flatten()
```

```python
#Plot AB40 to 10074G5
plt.plot(Time, AB40_to_10074G5)
plt.title("AB40 to 10074G5")
plt.xlabel("Time (mins)")
plt.ylabel("Power (ucal/sec)")
plt.show()
```

```python
#Plot AB40 to buffer
plt.plot(Time, AB40_to_buffer)
plt.title("AB40 to buffer")
plt.xlabel("Time (mins)")
plt.ylabel("Power (ucal/sec)")
plt.show()
```

```python
#Plot Buffer to 10074G5
plt.plot(Time, Buffer_to_10074G5)
plt.title("Buffer to 10074G5")
plt.xlabel("Time (mins)")
plt.ylabel("Power (ucal/sec)")
plt.show()
```

```python
#Plot AB40 to 10074G5, accounting for heats of dilution
plt.plot(Time, AB40_to_10074G5-AB40_to_buffer-Buffer_to_10074G5)
plt.title("AB40 to 10074G5, accounting for heats of dilution")
plt.xlabel("Time (mins)")
plt.ylabel("Power (ucal/sec)")
plt.show()
```

```python
#Plot integrated peaks, accounting for heats of dilution. 
#Please note that these were calculated elsewhere using the Origin.
fig, ax = plt.subplots()
plt.scatter(Molar_ratio, dQ/1000)
plt.title("")
plt.xlabel("Molar ratio")
plt.ylabel("$\Delta$ Q (kcal/mol)")
ax.set_ylim([-5,5])
plt.show()
```
