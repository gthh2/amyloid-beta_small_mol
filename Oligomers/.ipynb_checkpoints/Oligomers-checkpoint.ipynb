This notebook shows data from experiments involving the effects of 10074-G5 on stabilised oligomers of amyloid beta using dot blot analyses (using the OC and WO2 primary antibodies) and turbidimetry data collected at 600 nm. 

```python
#import useful things
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt; plt.rcdefaults()
from scipy import stats
%matplotlib inline
```

```python
#load in data

#turbidimetry data 
oligomers_0uM_10074G5  = pd.read_csv('Turbidimetry/10uM_oligomer_0uM_10074G5.csv',  header=None).values.flatten()
oligomers_20uM_10074G5 = pd.read_csv('Turbidimetry/10uM_oligomer_20uM_10074G5.csv', header=None).values.flatten()

#dot blot data
AB_alone_OC         = pd.read_csv('Dot_blot/AB_alone_OC_antibody.csv',         header=None).values.flatten()
AB_with_10074G5_OC  = pd.read_csv('Dot_blot/AB_with_10074G5_OC_antibody.csv',  header=None).values.flatten()

AB_alone_WO2        = pd.read_csv('Dot_blot/AB_alone_WO2_antibody.csv',        header=None).values.flatten()
AB_with_10074G5_WO2 = pd.read_csv('Dot_blot/AB_with_10074G5_WO2_antibody.csv', header=None).values.flatten()
```

```python
#plot turbidimetry data
vals = [1,2]
plt.bar(vals, [oligomers_0uM_10074G5.mean(), oligomers_20uM_10074G5.mean()], yerr=[oligomers_0uM_10074G5.std(), oligomers_20uM_10074G5.std()])

plt.title("Turbidimetry measurements")
plt.xlabel("10074-G5 concentration (uM)")
plt.ylabel("Absorbance (a.u.)")
plt.xticks(vals, ('0', '20'))

plt.show()
```

```python
#plot intensities from dot blot with the OC antibody 
vals = [1,2]
oc_vals = [AB_alone_OC.mean(), AB_with_10074G5_OC.mean()]
oc_errs = [AB_alone_OC.std(), AB_with_10074G5_OC.std()]
plt.bar(vals, oc_vals, yerr=oc_errs)

plt.title("Fluorescence intensity using OC antibody")
plt.xlabel("10074-G5 concentration (uM)")
plt.ylabel("Fluorescence intensity")
plt.xticks(vals, ('0', '20'))

plt.show()
```

```python
#plot intensities from dot blot with the WO2 antibody 

vals = [1,2]
wo2_vals = [AB_alone_WO2.mean(), AB_with_10074G5_WO2.mean()]
wo2_errs = [AB_alone_WO2.std(), AB_with_10074G5_WO2.std()]
plt.bar(vals, wo2_vals, yerr=wo2_errs)

plt.title("Fluorescence intensity using WO2 antibody")
plt.xlabel("10074-G5 concentration (uM)")
plt.ylabel("Fluorescence intensity")
plt.xticks(vals, ('0', '20'))

plt.show()
```
