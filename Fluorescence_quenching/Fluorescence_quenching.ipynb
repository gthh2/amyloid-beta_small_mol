This notebook displays the fluorescence quenching experiments of the aggregation of amyloid beta (AB) in the absence and presence the small molecule, 10074-G5. The aggregation of monomeric AB (5 uM) was monitored in the absence and presence of 5 uM 10074-G5.

```python
#import useful things
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt; plt.rcdefaults()
from matplotlib.pyplot import cm
%matplotlib inline
```

```python
#load in data
AB = pd.read_csv('5uMAB_0uMdrug.csv').values
G5 = pd.read_csv('5uMAB_5uMdrug.csv').values
```

```python
#plot the data for AB alone using color to indicate time
color=cm.rainbow(np.linspace(0,1,15))
for i,c in zip(range(0,15),color):
    plt.scatter(AB[:,0],AB[:,i+1], color=c, label=4*i)
plt.title("AB alone")
plt.xlabel("wavelength (nm)")
plt.ylabel("Fluorescence intensity (a.u.)")
plt.legend(loc="best")
plt.legend(title='Time (mins)')
plt.ylim(0, 1100)
plt.show
```

```python
#plot the data for AB+10074-G5 using color to indicate time
color=cm.rainbow(np.linspace(0,1,15))
for i,c in zip(range(0,15),color):
    plt.scatter(G5[:,0],G5[:,i+1], color=c, label=4*i)
plt.title("AB + 10074-G5")
plt.xlabel("wavelength (nm)")
plt.ylabel("Fluorescence intensity (a.u.)")
plt.legend(loc="right")
plt.legend(title='Time (mins)')
plt.ylim(0, 1100)
plt.show
```
